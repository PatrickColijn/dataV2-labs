# Challenge 1 - Tic Tac Toe

In this lab you will perform deep learning analysis on a dataset of playing [Tic Tac Toe](https://en.wikipedia.org/wiki/Tic-tac-toe).

There are 9 grids in Tic Tac Toe that are coded as the following picture shows:

![Tic Tac Toe Grids](tttboard.jpg)

In the first 9 columns of the dataset you can find which marks (`x` or `o`) exist in the grids. If there is no mark in a certain grid, it is labeled as `b`. The last column is `class` which tells you whether Player X (who always moves first in Tic Tac Toe) wins in this configuration. Note that when `class` has the value `False`, it means either Player O wins the game or it ends up as a draw.

Follow the steps suggested below to conduct a neural network analysis using Tensorflow and Keras. You will build a deep learning model to predict whether Player X wins the game or not.

## Step 1: Data Engineering

This dataset is almost in the ready-to-use state so you do not need to worry about missing values and so on. Still, some simple data engineering is needed.

1. Read `tic-tac-toe.csv` into a dataframe.
1. Inspect the dataset. Determine if the dataset is reliable by eyeballing the data.
1. Convert the categorical values to numeric in all columns.
1. Separate the inputs and output.
1. Normalize the input data.

```python
# Import libraries 
import pandas as pd
import numpy as np 
```

```python
# your code here
tictactoe = pd.read_csv('tic-tac-toe.csv')
tictactoe
```

```python
# Using SciKit Learn
from sklearn.preprocessing import LabelEncoder

columns = ['TL', 'TM', 'TR', 'ML', 'MM', 'MR', 'BL', 'BM', 'BR']

# Encoding make column using LabelEncoder
for column in columns:
    labelencoder = LabelEncoder()
    tictactoe[column] = labelencoder.fit_transform(tictactoe[column])

# Viewing few rows of make and its encoded columns
tictactoe.sample(20)
```

```python
list(labelencoder.classes_)
```

```python
tictactoe['class'] = tictactoe['class'].astype(int)
tictactoe

import tensorflow as tf

tictactoe_norm = tf.keras.utils.normalize(tictactoe[columns], axis=1)
tictactoe_norm = tictactoe_norm.join(tictactoe['class'].astype(int))
tictactoe_norm
```

```python
from sklearn.preprocessing import OneHotEncoder

columns = ['TL', 'TM', 'TR', 'ML', 'MM', 'MR', 'BL', 'BM', 'BR']

result_columns = []
for col in columns:
    result_columns.append('b'+col)
    result_columns.append('o'+col)
    result_columns.append('x'+col)

# Creating instance of one-hot-encoder
enc = OneHotEncoder(categories='auto') # drop first to decrease multicollinearity , drop='first' doesn't work(?)

# Encode values 
feature_arr = enc.fit_transform(tictactoe[columns]).toarray()
feature_labels = enc.categories_
feature_labels = np.array(feature_labels).ravel()
encoded_df = pd.DataFrame(feature_arr, columns=result_columns) # Create dataframe again
encoded_df

# merge with class column
tictactoe_enc = encoded_df.join(tictactoe['class'].astype(int))
tictactoe_enc
```

## Step 2: Build Neural Network

To build the neural network, you can refer to your own codes you wrote while following the [Deep Learning with Python, TensorFlow, and Keras tutorial](https://www.youtube.com/watch?v=wQ8BIBpya2k) in the lesson. It's pretty similar to what you will be doing in this lab.

1. Split the training and test data.
1. Create a `Sequential` model.
1. Add several layers to your model. Make sure you use ReLU as the activation function for the middle layers. Use Softmax for the output layer because each output has a single lable and all the label probabilities add up to 1.
1. Compile the model using `adam` as the optimizer and `sparse_categorical_crossentropy` as the loss function. For metrics, use `accuracy` for now.
1. Fit the training data.
1. Evaluate your neural network model with the test data.
1. Save your model as `tic-tac-toe.model`.

```python
from sklearn.model_selection import train_test_split

# Split the data 80-20 for training - test purposes
X = tictactoe_enc.drop("class",axis=1).to_numpy() 
y = tictactoe_enc["class"].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
X_train.shape
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(27,)))
model.add(tf.keras.layers.Dense(560, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(560, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(2, activation=tf.nn.sigmoid))

model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

## Step 3: Make Predictions

Now load your saved model and use it to make predictions on a few random rows in the test dataset. Check if the predictions are correct.

```python
# your code here
val_loss, val_acc = model.evaluate(X_test, y_test)
```

```python
# Save the model 
model.save('tictactoe_model.h5')
```

```python
# Load model
tictactoe_model = tf.keras.models.load_model('tictactoe_model.h5')
```

```python
predictions = tictactoe_model.predict([X_test])
print(np.argmax(predictions[0]))
```
