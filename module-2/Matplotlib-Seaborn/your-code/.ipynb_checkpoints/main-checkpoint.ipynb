#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
from math import e
x = np.arange(0,100)
y1 = x ** 2
y2 = e ** x

fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(8, 8))
ax1.plot(x, y1)
ax1.set_title('y=x2 PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')

ax2.plot(x, y2)
ax2.set_title('y=e^x PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_xscale('log')
ax2.set_yscale('log')

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y1 = x ** 2
y2 = e ** x

fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(8, 12))
ax1.plot(x, y1, label = 'test1')
ax1.set_title('y=x2 PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')
ax1.set_xscale('linear')
ax1.set_yscale('linear')
ax1.legend()

ax2.plot(x, y2, label = 'test2')
ax2.set_title('y=e^x PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.legend()

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

listofplots = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

plt.figure(figsize=(20,10)) # Not in for loop otherwise you will create a new figure for every plot 

for plot in listofplots: 
    x = fitbit[plot]
    y = fitbit['Steps']
    plt.scatter(x,y, label = plot)
    plt.xlabel('Minutes')
    plt.ylabel('Steps')
    plt.legend()
    
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.

```python
# Method 1 - matplotlib
sex_count = cat.groupby('Sex').count()
sex_count = sex_count.reset_index()
```

```python
labels = ['Male', 'Female']
sizes = (314, 573)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
# Method 2 - seaborn

fig, ax = plt.subplots(figsize = (12,6))
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#909090'
plt.rcParams['axes.labelcolor']= '#909090'
plt.rcParams['xtick.color'] = '#909090'
plt.rcParams['ytick.color'] = '#909090'
plt.rcParams['font.size']=12
ind = 2
color_palette_list = ['#009ACD', '#ADD8E6', '#63D1F4', '#0EBFE9', 
                      '#C1F0F6', '#0099CC']
sns.barplot(x=sex_count['Name'], y = sex_count['Sex'],  
            data = sex_count, palette=color_palette_list, 
            label="Number of people per gender", 
            ax=ax, ci=None)
ax.set_title("Elephant in the Valley Survey Results")
ax.set_ylabel("Percentage of Yes Responses to Question")
ax.set_ylim(0,2)
ax.set_xlabel("Question Key")
ax.set_xticks(range(0, ind))
ax.set_xticklabels((0, 1), rotation=180)
```
