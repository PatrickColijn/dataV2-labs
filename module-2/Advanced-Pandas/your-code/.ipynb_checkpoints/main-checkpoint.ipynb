Now create the standardized columns `CGPA_std`, `GRE_std`, and `LOR_std` and add them to the `admissions` dataframe.

```python
# your code here
admissions['CGPA_std'] = admissions[['CGPA']].apply(lambda col : (col - np.mean(col)) / np.std(col))
admissions['GRE_std'] = admissions[['GRE Score']].apply(lambda col : (col - np.mean(col)) / np.std(col))
admissions['LOR_std'] = admissions[['LOR']].apply(lambda col : (col - np.mean(col)) / np.std(col))
admissions.head()
```

We will generate the decision choice at random using the code below. Please run the cell.

```python
# Libraries
from random import choices
```

```python
std_columns = ['CGPA_std', 'GRE_std', 'LOR_std']

decision_choice = choices(std_columns, k=admissions.shape[0])
```

Now create the deciding column using the `lookup` function. The lookup column is `decision_choice` found above. Call the column resulting from the lookup function `deciding_column` and add it to the `admissions` dataframe.

```python
# your code here
admissions['deciding_column'] = decision_choice
#admissions.drop('decision_choice', axis=1, inplace=True)
admissions
```

Create a column called `decision` in the `admissions` dataframe. Assign 1 to this column if the value of `deciding_column` is greater than 0.8 and 0 otherwise.

```python
# TA'S: CAN WE DISCUSS THIS ONE PLEASE, THANK YOU! 


#admissions['decision'] = admissions[['deciding_column']].apply(lambda x : admissions[[x]])
#admissions
#admissions['decision'] = admissions[['deciding_column']].apply(lambda str(row) : 1 if admissions[[str(row)]] > 0.8 else 0)

criteria_GRE = np.where((admissions['GRE_std'] > 0.8),1,0)
criteria_CGPA = np.where((admissions['CGPA_std'] > 0.8),1,0)
criteria_LOR = np.where((admissions['LOR_std'] > 0.8),1,0)
admissions["decision"] = np.where(admissions['deciding_column'] == "GRE_std",criteria_GRE, \
                                  np.where(admissions['deciding_column'] == "CGPA_std", criteria_CGPA,
                                           np.where(admissions['deciding_column'] == "LOR_std", criteria_LOR, 0)))
admissions
```
