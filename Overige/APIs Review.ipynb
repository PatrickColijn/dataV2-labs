```python
# Import libraries
import pandas as pd
import requests
import json
```

```python
url = 'https://cat-fact.herokuapp.com/facts'
res = requests.get(url)
data = res.json() 
```

```python
# Create a dataframe of the json file 
df = pd.DataFrame(data['all'])
df.head()
```

```python
from pandas.io.json import json_normalize
flattened_data = json_normalize(data['all'])
flattened_data
#print(flattened_data['text'][0:10])
```

```python
import random

def random_catfact(data):
    print(data['text'][random.randint(0, len(data))])

random_catfact(flattened_data)
```
