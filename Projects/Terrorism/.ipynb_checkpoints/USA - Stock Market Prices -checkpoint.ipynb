# USA - Stock Market Prices

We will be looking at the three most influencial stockindexes of the US to determine the impact of terrorism on the US stock market. 
* The S&P 500
* The Dow Jones Industrial Average (DJIA)
* The Nasdaq Composite Index

```python
# Import Yahoo Finance library
import yfinance as yf 

# Assign tickers to retrieve stock data 
SP500 = yf.Ticker('^GSPC')
DowJones = yf.Ticker('^DJI')
Nasdaq = yf.Ticker('^IXIC')

# Retrieve stock data 
data_SP500 = SP500.history(period="max")
data_DowJones = DowJones.history(period="max")
data_Nasdaq = Nasdaq.history(period="max")
```

# First Stock Data Analysis
